## Set unique number for each label

```python
unique_labels = df.label.unique()
```

```python
unique_labels
```

```python
def getIndexFromList(rowToCheck):
  textToCheck = rowToCheck['label']
  for i in range(len(unique_labels)):
    if unique_labels[i] == textToCheck:
      return i
      break
  return -1
  
```

```python
#getIndexFromList('MAGNETIC SHOPPING ')
```

```python
df['NbLabel'] = df.apply (lambda row: getIndexFromList(row), axis=1)
```

```python
df.head(50)
```

## Set unique number for each general category of product

```python
unique_category = df.General.unique()

unique_category
```

```python
def getIndexFromListCateg(rowToCheck):
  textToCheck = rowToCheck['General']
  for i in range(len(unique_category)):
    if unique_category[i] == textToCheck:
      return i
      break
  return -1
```

```python
df['NbGeneralCategot'] = df.apply (lambda row: getIndexFromListCateg(row), axis=1)
```
